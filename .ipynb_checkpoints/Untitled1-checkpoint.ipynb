```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten,Dense
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
def random_points_generator(no_of_images):
    generated_images=np.random.uniform(size=(no_of_images,28,28))
    return generated_images
```

```python
def generator(number,x_train=x_train,y_train=y_train):
    data_set=x_train[y_train==number]
    training_points=random_points_generator(len(data_set))
    model=Sequential()
    model.add(Flatten(input_shape=(28,28,1)))
    model.add(Dense(64,activation='relu'))
    model.add(Dense(32,activation='relu'))
    model.add(Dense(64,activation='relu'))
    model.add(Dense(28*28,activation='sigmoid'))
    
    model.compile(optimizer='adam',loss='mse',metrics=['Accuracy'],run_eagerly=True)
    history=model.fit(training_points,data_set.reshape(-1,28*28),epochs=10,batch_size=32)
    plt.plot(history.history['Accuracy'])
    return (model,training_points,data_set.reshape(-1,28*28))
```

```python
def discriminatior(generator,number,x_train=x_train,y_train=y_train):
    data_set=x_train[y_train==number]
    training_points=random_points_generator(len(data_set))
    generated_images=generator.predict(training_points)
    Gen=np.zeros(len(generated_images))
    Real=np.ones(len(data_set))
    y_pre=np.concatenate((Gen,Real)).reshape(-1,1)
    x_tr=np.concatenate((generated_images,data_set.reshape(-1,28*28)))
    model=Sequential()
    model.add(Dense(64,activation='relu',input_shape=(28*28,)))
    model.add(Dense(32,activation='relu'))
    model.add(Dense(64,activation='relu'))
    model.add(Dense(1,activation='sigmoid'))
    model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['Accuracy'],run_eagerly=True)
    history=model.fit(x_tr,y_pre,epochs=10,batch_size=32)
    plt.plot(history.history['Accuracy'])
    return model
```

```python
Generator=generator(9)
Discriptor=discriminatior(Generator[0],9)
for i in range(3):
    prediction=Generator[0].predict(random_points_generator(25))
    for i in range(25):
        plt.subplot(5,5,i+1)
        plt.imshow(prediction[i].reshape(28,28))
    plt.show()
    ans=Discriptor.predict(prediction)
    if np.sum(ans==1)>np.sum(ans==0):
        Discriptor=discriminatior(Generator,9)
    else:
        Generator[0].fit(Generator[1],Generator[2],epochs=10,batch_size=32)
```
